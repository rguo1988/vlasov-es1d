```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
NL = 100
NH = 50
L = 10
a = 5
TH = 10
TL = 1
EP = 0.1
Ta = (NL*TL+NH/L*a*TH)/(NL+NH*a/L)
Tna = TH-EP
Na = NL/a+NH/L
Nna = NH/L
T_aver = (NH*TH*a/L+NH*(TH-EP)*(L-a)/L+NL*TL)/(NH+NL)
n_aver = (NH+NL)/L
```

```python
nx = 101
x = np.linspace(0,L,num=nx,endpoint=True)
dx = x[1]-x[0]
a_start = int((L/2-a/2)/dx)
a_end = int((L/2+a/2)/dx)
T = Tna * np.ones(nx)
T[a_start:a_end]=Ta
n = Nna * np.ones(nx)
n[a_start:a_end]=Na
fig = plt.figure(figsize=(12,6))
ax1 = plt.subplot(121)
ax1.plot(x,T,color='red')
ax1.axhline(T_aver)
ax2 = plt.subplot(122)
ax2.plot(x,n,color='black')
ax2.axhline(n_aver)
plt.show()
```

```python
plt.plot(x,np.log(n/n_aver)/np.log(T/T_aver))
plt.show()
```
